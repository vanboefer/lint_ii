```python
from lint_ii import WordFeatures, SentenceAnalysis, ReadabilityAnalysis

import json
from pathlib import Path

DATA = Path('../../data')
example_texts = json.loads((DATA / 'examples_for_visualization.json').read_text(encoding='utf8'))
example_texts
```

```python
for example in example_texts:
    fp = DATA / f"{example['cefr_level'][:2]}.json"
    ra = ReadabilityAnalysis.from_text(example['text'])
    dump = json.dumps(ra.as_dict())
    fp.write_text(dump, encoding='utf8')
```
